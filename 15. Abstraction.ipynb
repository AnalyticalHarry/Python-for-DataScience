    <h2>Four Pillars of Object-Oriented Programming
</h2>
</div>


Object-Oriented Programming (OOP) is built on four fundamental principles, often referred to as the pillars of OOP. These principles shape the foundation of OOP by providing a structured and organized approach to software development.

### 1. Encapsulation

Encapsulation is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit known as a class. This concept hides the internal state of an object from the outside and exposes a controlled interface to interact with it. It ensures that the data within an object is kept safe from outside interference and misuse. Access to the data is typically controlled through methods such as getters and setters provided by the class. Encapsulation promotes data security, integrity, and reusability of code.

### 2. Abstraction

Abstraction involves simplifying complex systems by providing a simplified interface. It focuses on showing only essential details and hiding the implementation details. Through abstraction, classes and objects provide a way to represent the structure and behavior without exposing all the underlying complexities. It enables programmers to work with high-level concepts without worrying about the intricacies of the internal code. For instance, a car object might have methods like `start()` and `stop()`, allowing users to interact without knowing the detailed mechanics of these functions.

### 3. Inheritance

Inheritance is a mechanism that allows new classes (derived or child classes) to be created based on existing classes (base or parent classes), inheriting their attributes and methods. This principle fosters code reusability by enabling new classes to leverage the properties and behavior of existing classes. The relationship established through inheritance forms an "is-a" connection, where the derived class is a specialized version of the base class. It reduces redundancy in code and helps in building a hierarchy of classes.

### 4. Polymorphism

Polymorphism allows objects to be treated as instances of their parent class, enabling them to be used interchangeably. It's achieved through method overriding and method overloading. Method overriding allows a subclass to provide a specific implementation of a method that is already defined in its parent class. Method overloading involves having multiple methods with the same name but different parameters within the same class. Polymorphism provides flexibility in the use of objects and methods, making the code more adaptable and scalable.

These four pillars of OOP collectively provide a powerful framework for designing and building robust, maintainable, and scalable software systems.

#### Abstraction in Frameworks

Abstraction in the context of a framework refers to the concept of hiding complex implementation details and emphasizing only the essential aspects or functionalities to the users or developers interacting with that framework. It involves creating a simplified model or interface that allows users to interact with the framework without needing to understand the intricate inner workings.

Frameworks often consist of various layers of abstraction that help in managing complexity and making it easier for developers to use the provided tools or functionalities. This abstraction is achieved through several mechanisms:

1. **Interfaces and APIs:** Application programming interfaces (APIs) that expose specific functionalities or services while hiding the underlying complexity. This allows developers to interact with the framework using predefined methods, functions, or classes without needing to understand the intricate details of how those methods are implemented.

2. **Encapsulation:** Abstraction also involves encapsulating the implementation details within the framework components. Users can interact with these components without needing to know how they are built or how they function internally. This encapsulation helps in maintaining a clear boundary between different components, making it easier to use them independently.

3. **Simplified Models:** For instance, in a web development framework, developers can use high-level functions or classes to handle HTTP requests without having to deal with the low-level networking intricacies.

4. **Hiding Complexity:** The abstraction in a framework hides the complexity of the underlying system. This allows developers to focus on utilizing the provided functionalities without worrying about the complex algorithms, data structures, or low-level operations.


#### Benefits of `@abstractmethod` 

The `@abstractmethod` decorator from the `abc` (Abstract Base Classes) module in Python offers several advantages when defining abstract methods within abstract classes:

1. **Forcing Implementation in Subclasses:**
   - *Enforcement of Overriding:* It ensures that any concrete subclass derived from an abstract class must provide an implementation for the abstract method, preventing accidental omission of required functionality in subclasses.

2. **Consistent Interface:**
   - *Maintaining a Common Structure:* It fosters a consistent interface for related classes, ensuring that all derived classes share the same method signature while allowing each class to implement the method according to its specific behavior.

3. **Enhancing Code Readability and Usability:**
   - *Clarifies Intent:* It explicitly marks methods as abstract, indicating to developers that these methods need implementation in derived classes.
   - *Improves Maintainability:* Facilitates code understanding and maintenance by clearly defining expectations for subclass implementation.

4. **Polymorphism and Runtime Behavior:**
   - *Support for Polymorphism:* Enables polymorphic behavior, allowing different objects to be treated uniformly based on their common abstract interface.
   - *Run-time Checks:* When an abstract method is called on an instance and not implemented in the specific class, the interpreter raises an error at runtime, highlighting the need for method implementation.

5. **Designing Hierarchies and Protocols:**
   - *Structuring Class Hierarchies:* Facilitates the design of class hierarchies, particularly in cases where specific methods need to be implemented in child classes while sharing common signatures.

6. **Improving Code Robustness and Preventing Bugs:**
   - *Early Identification of Missing Implementations:* Helps detect missing implementations during development or testing, preventing runtime errors and potential bugs due to incomplete functionality.

Overall, the `@abstractmethod` decorator provides structure and guidelines for class hierarchies, ensuring a consistent and predictable implementation across different subclasses. It offers benefits in terms of readability, maintainability, and robustness in the design and implementation of object-oriented systems.


```python
from abc import ABC, abstractmethod
```

```python
#abstract class 'Shape' representing a generic shape
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass

#concrete class 'Square' inheriting from 'Shape'
class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    # This class is forced to implement the abstract method 'calculate_area'
    def calculate_area(self):
        return self.side_length * self.side_length
```

```python
#instantiating and using the classes
square = Square(5)
print("Area of the square:", square.calculate_area())
```

```python
#abstract class 'Shape' without @abstractmethod decorator
class Shape(ABC):
    def calculate_area(self):
        pass

#concrete class 'Square' inheriting from 'Shape'
class Square(Shape):
    def __init__(self, side_length):
        self.side_length = side_length

    #although 'calculate_area' is defined in the abstract class, it's not enforced to be implemented
    def calculate_area(self):
        return self.side_length * self.side_length
```

```python
#instantiating and using the classes
square = Square(5)
print("Area of the square:", square.calculate_area())
```

#### Abstract Base Class for Shapes

The code defines a base class called 'Shape'. This class is designed to serve as an abstract representation for various shapes.

#### Methods within the 'Shape' Class:

- **area():** This method is declared as an abstract method using the `pass` statement. It's intended to calculate the area of a shape. However, as an abstract method, it lacks an implementation in the base class. The idea is that any subclass of 'Shape' must provide its own implementation of the `area()` method to calculate the area specific to that shape.

- **perimeter():** Similar to the `area()` method, this method is also declared as an abstract method using `pass`. It is meant to calculate the perimeter (or circumference in some cases) of a shape. As an abstract method, it doesn't contain the specific logic to compute the perimeter. Instead, any subclass of 'Shape' is required to implement its own `perimeter()` method to calculate the shape's perimeter.

#### Purpose:

The purpose of these abstract methods within the 'Shape' class is to provide a blueprint or interface that any specific shape subclass must adhere to. By defining these abstract methods in the base class, it enforces that any subclass (e.g., 'Rectangle', 'Circle', etc.) must implement these methods with their own custom logic for area and perimeter calculations based on the shape's specific characteristics.


```python
import math
```

```python
math.pi
```

```python
#define a base class called 'Shape'
class Shape:
    #abstract method for calculating the area of a shape
    def area(self):
        pass
    #abstract method for calculating the perimeter of a shape
    def perimeter(self):
        pass
```

#### 'Square' Class Inheriting from 'Shape'

The code defines a class named 'Square' that inherits from the 'Shape' class, thereby extending and implementing the abstract methods for calculating area and perimeter specifically for squares.

#### Methods and Initialisation in the 'Square' Class:

- **__init__(self, side_length):** This method is the constructor used to initialize objects of the 'Square' class. It takes a parameter `side_length` and sets the `self.side_length` attribute, representing the side length of the square.

- **area(self):** This method overrides the abstract `area()` method inherited from the 'Shape' class. It provides a specific implementation for calculating the area of a square. In this case, it multiplies the side length by itself (`side_length * side_length`) to compute the area.

- **perimeter(self):** Similar to the `area()` method, this method overrides the abstract `perimeter()` method inherited from the 'Shape' class. It offers a specific implementation for calculating the perimeter of a square. It computes the perimeter by multiplying the side length by 4 (`4 * side_length`), as a square has four equal sides.

#### Purpose:

The 'Square' class exemplifies the concept of inheritance and polymorphism by inheriting the 'area()' and 'perimeter()' abstract methods from the abstract 'Shape' class and providing its own specific implementations tailored for calculating the area and perimeter of a square. This setup allows 'Square' objects to be created, leveraging the methods defined in the 'Shape' class, yet providing specialized behavior for calculating area and perimeter unique to squares.


```python
#define a class 'Square' that inherits from 'Shape'
class Square(Shape):
    #initialise with the side length of the square
    def __init__(self, side_length):
        self.side_length = side_length

    #method to calculate the area of a square
    def area(self):
        return self.side_length * self.side_length

    #method to calculate the perimeter of a square
    def perimeter(self):
        return 4 * self.side_length
```

#### 'Circle' Class Inheriting from 'Shape'

The code introduces a class named 'Circle' that inherits from the 'Shape' class. This class provides specific implementations for calculating the area and perimeter of a circle.

#### Methods and Initialisation in the 'Circle' Class:

- **__init__(self, radius):** This method serves as the constructor for 'Circle' objects. It initializes the radius of the circle when a 'Circle' object is created.

- **area(self):** This method overrides the abstract `area()` method inherited from the 'Shape' class. It offers a specific implementation for calculating the area of a circle. The area is calculated using the formula:  * r, where 'r' is the radius of the circle.

- **perimeter(self):** Similar to the `area()` method, this method overrides the abstract `perimeter()` method inherited from the 'Shape' class. It provides a specific implementation to calculate the circumference (or perimeter) of a circle. The formula used here is: 2 *  * r, where 'r' represents the radius of the circle.

**Note:** The code snippet assumes the availability of the 'math' module for using the constant 'pi' to approximate the value of  for area and perimeter calculations.

#### Purpose:

The 'Circle' class demonstrates inheritance and polymorphism by inheriting the abstract methods for area and perimeter calculations from the 'Shape' class and providing specialized methods tailored for calculating the area and perimeter of circles. This enables 'Circle' objects to leverage the methods defined in the 'Shape' class while offering specific implementations customized for circles.


```python
#define a class 'Circle' that inherits from 'Shape'
class Circle(Shape):
    #initialise with the radius of the circle
    def __init__(self, radius):
        self.radius = radius

    #method to calculate the area of a circle
    def area(self):
        return math.pi * self.radius * self.radius  #using an approximation of pi

    #method to calculate the perimeter (circumference) of a circle
    def perimeter(self):
        return 2 * math.pi * self.radius  #using an approximation of pi for circumference
```

```python
square = Square(5)
circle = Circle(7)
```

```python
print("Square Area:", square.area())
print("Square Perimeter:", square.perimeter())
```

```python
print("Circle Area:", circle.area())
print("Circle Perimeter:", circle.perimeter())
```

#### Concrete Classes 'Television' and 'Laptop'
- `Television` and `Laptop` are concrete classes that inherit from the abstract class `ElectronicDevice`.
- These classes provide specific implementations for the abstract methods `power_on()` and `power_off()`, defining the behavior for turning on and off their respective devices.

#### Usage of the Classes
Within the `if __name__ == "__main__":` block:
- Instances of `Television` and `Laptop` are created.
- The code demonstrates the usage of the `power_on()` and `power_off()` methods for both a television and a laptop, showcasing the specific behavior implemented in each concrete class.

```python
from abc import ABC, abstractmethod
```

```python
#abstract class 'ElectronicDevice' representing a generic electronic device
class ElectronicDevice(ABC):
    @abstractmethod
    def power_on(self):
        pass

    @abstractmethod
    def power_off(self):
        pass
```

```python
#concrete class 'Television' inheriting from 'ElectronicDevice'
class Television(ElectronicDevice):
    def power_on(self):
        return "Television powered on."

    def power_off(self):
        return "Television powered off."
```

```python
#concrete class 'Laptop' inheriting from 'ElectronicDevice'
class Laptop(ElectronicDevice):
    def power_on(self):
        return "Laptop powered on."

    def power_off(self):
        return "Laptop powered off."
```

```python
#using the classes
if __name__ == "__main__":
    tv = Television()
    print(tv.power_on())
    print(tv.power_off())
```

```python
#using the classes
if __name__ == "__main__":
    laptop = Laptop()
    print(laptop.power_on())
    print(laptop.power_off())
```

```python
from abc import ABC, abstractmethod
```

```python
#abstract class 'Vehicle' representing a generic vehicle
class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

#concrete class 'Car' inheriting from 'Vehicle'
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started."

    def stop_engine(self):
        return "Car engine stopped."

#concrete class 'Motorcycle' inheriting from 'Vehicle'
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started."

    def stop_engine(self):
        return "Motorcycle engine stopped."
```

```python
#using the classes
if __name__ == "__main__":
    my_car = Car()
    print(my_car.start_engine())
    print(my_car.stop_engine())
```

```python
#using the classes
if __name__ == "__main__":
    my_motorcycle = Motorcycle()
    print(my_motorcycle.start_engine())
    print(my_motorcycle.stop_engine())
```

</h2>

Abstraction in a scenario of calculating salaries for different types of employees.

#### Abstract Class 'Employee'

- The abstract class `Employee`, defined as an abstract base class (ABC), represents a generic employee.
- It includes an abstract method `calculate_salary()`, which acts as a placeholder method without any implementation details. Subclasses are obligated to override this method.

#### Concrete Subclasses - Developer, Manager, Intern

- **Developer Class:**
  - Inherits from `Employee` and provides a specific implementation for `calculate_salary()` method.
  - Represents a scenario where developers have a fixed annual salary.

- **Manager Class:**
  - Inherits from `Employee` and offers an individualized implementation for `calculate_salary()` method.
  - Demonstrates a situation where managers might receive a base salary along with bonuses.

- **Intern Class:**
  - Inherits from `Employee` and provides a distinct implementation for `calculate_salary()` method.
  - Represents interns receiving a fixed stipend annually.

#### Usage of Classes

- Instances of `Developer`, `Manager`, and `Intern` are created, each initialized with a name and an employee ID.
- The `calculate_salary()` method is invoked for each instance to display the calculated salary or stipend based on their role.

#### Purpose and Significance

This code displays the concept of abstraction by defining an abstract base class `Employee` with an abstract method, ensuring specific behaviors for salary calculation are implemented in the subclasses. This allows different employee types to have their custom salary calculation methods while adhering to a common interface defined by the abstract class.

```python
from abc import ABC, abstractmethod
```

```python
#abstract class 'Employee' representing a generic employee
class Employee(ABC):
    def __init__(self, name, employee_id):
        self.name = name
        self.employee_id = employee_id

    @abstractmethod
    def calculate_salary(self):
        pass
```

```python
#concrete class 'Developer' inheriting from 'Employee'
class Developer(Employee):
    def calculate_salary(self):
        #for example, developers get a fixed salary or hourly rate
        return 70000  #return a fixed annual salary
```

```python
#concrete class 'Manager' inheriting from 'Employee'
class Manager(Employee):
    def calculate_salary(self):
        #for instance, managers might have a base salary plus bonuses
        return 90000  #return a fixed annual salary + bonuses
```

```python
#concrete class 'Intern' inheriting from 'Employee'
class Intern(Employee):
    def calculate_salary(self):
        #interns might have a stipend or no salary
        return 15000  #return a fixed annual stipend
```

```python
#creating instances of Developer, Manager, and Intern
developer = Developer("Alice", 101)
manager = Manager("Bob", 202)
intern = Intern("Eve", 303)
```

```python
#calculate_salary method
print(f"{developer.name}'s salary: ${developer.calculate_salary()} per year")
print(f"{manager.name}'s salary: ${manager.calculate_salary()} per year")
print(f"{intern.name}'s stipend: ${intern.calculate_salary()} per year")
```

#### Example 5: Finance Analaysis using Concept of Abstraction

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from abc import ABC, abstractmethod
```

```python
class Stock(ABC):
    def __init__(self, ticker, period):
        self.ticker = ticker
        self.period = period
    
    @abstractmethod
    def get_data(self):
        pass

class Price(Stock):
    def get_data(self):
        return yf.download(self.ticker, period=self.period)

class Chart:
    def plot(self, data):
        plt.figure(figsize=(15,4))
        plt.stackplot(data.index, data['Close'], alpha=0.9)
        plt.xlabel('Date')
        plt.ylabel('Close Price $')
        plt.grid(True, ls='--', color='black', alpha=0.3)
        plt.show()
```

```python
microsoft = "msft"
data_period = "5y"
```

```python
#an instance of the Price class
price_instance = Price(microsoft, data_period)
stock_data = price_instance.get_data()
```

```python
#an instance of the chart class and visualising the data
chart_instance = Chart()
chart_instance.plot(stock_data)
```
